```python
"""STEP 2 modification of initial budget file code"""

import pandas as pd
import numpy as np


budget = pd.read_excel('data_files/budget_file_intermediate.xlsx', sheet_name='budget_by_month')

def year(x): 
    y = str(x)
    yr = y[0:4]
    return yr

def months(x): 
    y = str(x)
    m = y[4:]
    return m

budget_w = budget.loc[:,['ID','OPCO', 'Tier', 'ONE_NAME', 'Indirect/Direct',
       'Channel', 'Type', 'OI Bgt FY21', 'OI Bgt FY22']]
budget_w['day'] = str(1)
df_2021 = budget_w.loc[:,['ID','OPCO', 'Tier', 'ONE_NAME', 'Indirect/Direct',
       'Channel', 'Type', 'OI Bgt FY21']]
df_2021['FY'] = str(2021)
df_2022 = budget_w.loc[:,['ID','OPCO', 'Tier', 'ONE_NAME', 'Indirect/Direct',
       'Channel', 'Type', 'OI Bgt FY22']]
df_2022['FY'] = str(2022)
df_2021.rename(columns = {'OI Bgt FY21' : 'Bgt'}, inplace = True)
df_2022.rename(columns = {'OI Bgt FY22' : 'Bgt'}, inplace = True)

df_1 = pd.DataFrame()
df_2 = pd.DataFrame()
lst_1 = []
lst_2 = []

for i in [202104, 202105, 202106, 202107, 202108, 202109, 202110, 202111,
       202112, 202201, 202202, 202203]:
    df_1 = df_2021.copy()
    df_1['year_month'] = str(i)
    lst_1.append(df_1)

for y in [202204, 202205, 202206, 202207,
       202208, 202209, 202210, 202211, 202212, 202301, 202302, 202303]:
    df_2 = df_2022.copy()
    df_2['year_month'] = str(y)
    lst_2.append(df_2)

lst = lst_1 + lst_2

df = pd.concat(lst, ignore_index = True)

#df = df[df['Bgt'] != 0]
#df.dropna(subset = 'Bgt',inplace = True)
df['Bgt'] = df['Bgt'] / 12
df['Bgt'] = df['Bgt'].round(decimals = 2)

years = df['year_month'].apply(lambda x: year(x))
month = df['year_month'].apply(lambda x: months(x))
df.insert(5,'year', years)
df.insert(5,'month', month)

#df['date'] = df['month'] + '-' + df['day'] + '-' + df['year']
#df['date'] = df['date'].astype('datetime64')

#df.rename(columns = {'sold_to_customer_0' : 'sold_to_customer', 'Customer Name' : 'sold_to_customer_n'}, inplace = True)
df = df.loc[:,['ID', 'OPCO', 'Tier', 'ONE_NAME', 'Indirect/Direct',
       'Channel', 'Type', 'Bgt', 'year_month', 'FY']]
#df = df.set_index('ID')

df.ID = df.ID.astype('str')
df.ID = "0" + df.ID
writer = pd.ExcelWriter('data_files/budget_file_for_tableau.xlsx', engine='xlsxwriter')
df.to_excel(writer, sheet_name='budget_by_month')
writer.save()
```
